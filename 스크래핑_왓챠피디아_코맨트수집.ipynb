# 왓챠피디아 코멘트 스크래핑 도구 (Selenium & Colab)

## 1. 이 코드는 무엇인가요?
이 코드는 **왓챠피디아(Watcha Pedia)** https://pedia.watcha.com/ko-KR 의 특정 영화나 드라마 페이지에 있는 모든 코멘트(댓글)를 자동으로 수집하여 `json` 파일로 저장해주는 도구입니다.

- **작동 방식**: 가상의 웹 브라우저(Chrome)를 실행시켜, 실제 사용자가 스크롤을 내리는 것처럼 페이지의 모든 댓글을 로딩한 후, ID, 별점, 내용, 좋아요 수, 댓글 수를 추출합니다.
- **특징**: 스포일러가 포함된 댓글의 '보기' 버튼을 자동으로 클릭하여 숨겨진 내용까지 모두 수집할 수 있습니다.
- **비고**: 네이버 영화 서비스 종료 후, 관객 반응 수집에 가장 적절한 플랫폼으로 생각됨.

````python
# 본 예시는 슈퍼맨(2025)로 코멘트가 비교적 적음. 대략 30초 이내 수집 가능.
# https://pedia.watcha.com/ko-KR/contents/md6lQ90/comments

import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

def get_all_comments_with_selenium(content_id, timeout_minutes=0):
    """
    Selenium을 사용해 웹 브라우저를 직접 제어하여 모든 코멘트를 스크래핑합니다.
    (Google Colab 환경에 최적화되었습니다.)
    """
    # --- ★★★ 중요: 스크래핑할 페이지의 CSS 선택자 ★★★ ---
    # 각 페이지마다 이 클래스 이름이 다를 수 있습니다.
    # 문제가 발생하면, 브라우저의 '검사(Inspect)' 도구로 이 값을 확인하고 수정해야 합니다.
    COMMENT_ITEM_SELECTOR = 'article.Odsjukuj'  # 댓글 하나하나를 감싸는 항목
    USER_ID_SELECTOR = 'p.KI8y5yor'             # 사용자 ID
    RATING_SELECTOR = 'div.GMsh784C p'         # 별점 (별점이 없는 경우도 처리)
    TEXT_SELECTOR = 'p.CommentText'            # 댓글 내용
    ACTION_BUTTON_SELECTOR = 'button.ygosCx8o' # '좋아요', '댓글' 버튼
    SPOILER_BUTTON_SELECTOR = 'button.eIu73FdM' # (추가됨) 스포일러 '보기' 버튼
    # ---------------------------------------------------------

    comments_data = []
    url = f"https://pedia.watcha.com/ko-KR/contents/{content_id}/comments"

    chrome_options = Options()
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")

    driver = None
    try:
        print("가상 웹 브라우저(Chrome)를 실행합니다...")
        driver = webdriver.Chrome(options=chrome_options)
        driver.get(url)

        print("댓글 목록이 나타날 때까지 최대 20초간 기다립니다...")
        WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, COMMENT_ITEM_SELECTOR))
        )
        print("댓글 목록을 확인했습니다. 페이지를 스크롤합니다...")

        start_time = time.time()
        last_height = driver.execute_script("return document.body.scrollHeight")

        while True:
            if timeout_minutes > 0:
                elapsed_seconds = time.time() - start_time
                if elapsed_seconds > timeout_minutes * 60:
                    print(f"\n[알림] 설정된 최대 수집 시간({timeout_minutes}분)을 초과하여 스크롤을 중단합니다.")
                    break

            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                print("페이지의 끝에 도달했습니다. 스크롤을 종료합니다.")
                break
            last_height = new_height

        # --- (업그레이드) 스포일러가 포함된 모든 댓글의 '보기' 버튼 클릭 ---
        try:
            print("스포일러 댓글을 확인하고 전체 내용을 로딩합니다...")
            # 페이지에 있는 모든 스포일러 '보기' 버튼을 찾습니다.
            spoiler_buttons = driver.find_elements(By.CSS_SELECTOR, SPOILER_BUTTON_SELECTOR)
            if spoiler_buttons:
                print(f"{len(spoiler_buttons)}개의 스포일러 댓글을 발견했습니다. 전체 내용을 표시합니다.")
                for button in spoiler_buttons:
                    # JavaScript를 이용해 안정적으로 버튼을 클릭합니다.
                    driver.execute_script("arguments[0].click();", button)
                time.sleep(1) # 내용이 모두 표시될 때까지 잠시 대기
        except Exception as e:
            print(f"[경고] 스포일러 버튼 클릭 중 오류가 발생했습니다: {e}")
        # ---------------------------------------------------------

        print("페이지의 모든 정보를 분석합니다...")
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        comment_items = soup.select(COMMENT_ITEM_SELECTOR)

        if not comment_items:
            print("[오류] 코멘트를 찾지 못했습니다. 코드 상단의 COMMENT_ITEM_SELECTOR 값을 확인해주세요.")
            return []

        print(f"총 {len(comment_items)}개의 코멘트를 발견했습니다. 데이터 추출을 시작합니다.")

        for comment in comment_items:
            user_id_tag = comment.select_one(USER_ID_SELECTOR)
            user_id = user_id_tag.text.strip() if user_id_tag else 'N/A'

            rating_tag = comment.select_one(RATING_SELECTOR)
            rating = rating_tag.text.strip() if rating_tag else '평가 없음'

            text_tag = comment.select_one(TEXT_SELECTOR)
            # 스포일러 '보기' 버튼이 클릭된 후에는 전체 텍스트가 포함되므로, '보기'라는 단어만 제거합니다.
            text = text_tag.text.strip().replace('보기', '') if text_tag else ''

            action_buttons = comment.select(ACTION_BUTTON_SELECTOR)
            likes = '0'
            replies = '0'
            for button in action_buttons:
                if '좋아요' in button.text:
                    likes = button.find('span').text if button.find('span') else '0'
                elif '댓글' in button.text:
                    replies = button.find('span').text if button.find('span') else '0'

            comments_data.append({
                "ID": user_id,
                "별점": rating,
                "내용": text,
                "좋아요": likes,
                "댓글 수": replies
            })

    except TimeoutException:
        print(f"[오류] 페이지 로딩 시간(20초)이 초과되었습니다. URL을 확인하거나 코드 상단의 COMMENT_ITEM_SELECTOR ('{COMMENT_ITEM_SELECTOR}')가 올바른지 확인해주세요.")
    except Exception as e:
        print(f"오류가 발생했습니다: {e}")
    finally:
        if driver:
            print("작업을 완료하고 브라우저를 종료합니다.")
            driver.quit()

    return comments_data

def print_guide():
    """콘텐츠 ID를 찾는 방법을 안내하는 함수"""
    print("=" * 60)
    print("왓챠피디아 콘텐츠 ID를 찾는 방법")
    print("=" * 60)
    print("1. 왓챠피디아에서 코멘트를 수집하고 싶은 영화/드라마 페이지로 접속합니다.")
    print("2. 브라우저의 주소창을 확인합니다.")
    print("   예시: 슈퍼맨(2025)로 코멘트가 비교적 적음. 대략 30초 이내 수집 가능.")
    print("   예시: https://pedia.watcha.com/ko-KR/contents/md6lQ90/comments")
    print("3. 위 주소에서 'contents/' 와 '/comments' 사이에 있는")
    print("   영문과 숫자로 조합된 코드(예: md6lQ90)가 바로 '콘텐츠 ID' 입니다.")
    print("\n※ 안내: 코멘트가 많은 콘텐츠는 수집에 몇 분 이상 소요될 수 있습니다.")
    print("-" * 60)

if __name__ == "__main__":

    print_guide()

    content_id = ""
    while not content_id:
        content_id = input("▶ 수집할 콘텐츠의 ID를 입력하고 Enter 키를 누르세요: ").strip()
        if not content_id:
            print("[오류] ID가 입력되지 않았습니다. 다시 입력해주세요.")

    timeout_minutes = 0
    while True:
        timeout_input = input("▶ 최대 수집 시간을 분 단위로 입력하세요 (전체 수집은 0 또는 Enter): ").strip()
        if not timeout_input or timeout_input == "0":
            print("[알림] 시간 제한 없이 모든 댓글을 수집합니다.")
            break
        try:
            timeout_minutes = int(timeout_input)
            if timeout_minutes < 0:
                print("[오류] 0 또는 양수를 입력해주세요.")
            else:
                print(f"[알림] 최대 {timeout_minutes}분 동안 댓글을 수집합니다.")
                break
        except ValueError:
            print("[오류] 숫자만 입력해주세요.")

    comments = get_all_comments_with_selenium(content_id, timeout_minutes)

    if comments:
        print(f"\n총 {len(comments)}개의 코멘트를 성공적으로 수집했습니다.")
        file_name = f"watcha_comments_{content_id}.json"
        with open(file_name, "w", encoding="utf-8") as f:
            json.dump(comments, f, ensure_ascii=False, indent=4)
        print(f"'{file_name}' 파일로 저장이 완료되었습니다.")
    else:
        print("수집된 코멘트가 없습니다.")
````

````python
# 본 예시는 오징어 게임 시즌3(2025)로 코멘트가 비교적 많음. 대략 5분 이상 소요.
# hhttps://pedia.watcha.com/ko-KR/contents/tRMx0Jq/comments


import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

def get_all_comments_with_selenium(content_id, timeout_minutes=0):
    """
    Selenium을 사용해 웹 브라우저를 직접 제어하여 모든 코멘트를 스크래핑합니다.
    (Google Colab 환경에 최적화되었습니다.)
    """
    # --- ★★★ 중요: 스크래핑할 페이지의 CSS 선택자 ★★★ ---
    # 각 페이지마다 이 클래스 이름이 다를 수 있습니다.
    # 문제가 발생하면, 브라우저의 '검사(Inspect)' 도구로 이 값을 확인하고 수정해야 합니다.
    COMMENT_ITEM_SELECTOR = 'article.Odsjukuj'  # 댓글 하나하나를 감싸는 항목
    USER_ID_SELECTOR = 'p.KI8y5yor'             # 사용자 ID
    RATING_SELECTOR = 'div.GMsh784C p'         # 별점 (별점이 없는 경우도 처리)
    TEXT_SELECTOR = 'p.CommentText'            # 댓글 내용
    ACTION_BUTTON_SELECTOR = 'button.ygosCx8o' # '좋아요', '댓글' 버튼
    SPOILER_BUTTON_SELECTOR = 'button.eIu73FdM' # (추가됨) 스포일러 '보기' 버튼
    # ---------------------------------------------------------

    comments_data = []
    url = f"https://pedia.watcha.com/ko-KR/contents/{content_id}/comments"

    chrome_options = Options()
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")

    driver = None
    try:
        print("가상 웹 브라우저(Chrome)를 실행합니다...")
        driver = webdriver.Chrome(options=chrome_options)
        driver.get(url)

        print("댓글 목록이 나타날 때까지 최대 20초간 기다립니다...")
        WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, COMMENT_ITEM_SELECTOR))
        )
        print("댓글 목록을 확인했습니다. 페이지를 스크롤합니다...")

        start_time = time.time()
        last_height = driver.execute_script("return document.body.scrollHeight")

        while True:
            if timeout_minutes > 0:
                elapsed_seconds = time.time() - start_time
                if elapsed_seconds > timeout_minutes * 60:
                    print(f"\n[알림] 설정된 최대 수집 시간({timeout_minutes}분)을 초과하여 스크롤을 중단합니다.")
                    break

            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                print("페이지의 끝에 도달했습니다. 스크롤을 종료합니다.")
                break
            last_height = new_height

        # --- (업그레이드) 스포일러가 포함된 모든 댓글의 '보기' 버튼 클릭 ---
        try:
            print("스포일러 댓글을 확인하고 전체 내용을 로딩합니다...")
            # 페이지에 있는 모든 스포일러 '보기' 버튼을 찾습니다.
            spoiler_buttons = driver.find_elements(By.CSS_SELECTOR, SPOILER_BUTTON_SELECTOR)
            if spoiler_buttons:
                print(f"{len(spoiler_buttons)}개의 스포일러 댓글을 발견했습니다. 전체 내용을 표시합니다.")
                for button in spoiler_buttons:
                    # JavaScript를 이용해 안정적으로 버튼을 클릭합니다.
                    driver.execute_script("arguments[0].click();", button)
                time.sleep(1) # 내용이 모두 표시될 때까지 잠시 대기
        except Exception as e:
            print(f"[경고] 스포일러 버튼 클릭 중 오류가 발생했습니다: {e}")
        # ---------------------------------------------------------

        print("페이지의 모든 정보를 분석합니다...")
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        comment_items = soup.select(COMMENT_ITEM_SELECTOR)

        if not comment_items:
            print("[오류] 코멘트를 찾지 못했습니다. 코드 상단의 COMMENT_ITEM_SELECTOR 값을 확인해주세요.")
            return []

        print(f"총 {len(comment_items)}개의 코멘트를 발견했습니다. 데이터 추출을 시작합니다.")

        for comment in comment_items:
            user_id_tag = comment.select_one(USER_ID_SELECTOR)
            user_id = user_id_tag.text.strip() if user_id_tag else 'N/A'

            rating_tag = comment.select_one(RATING_SELECTOR)
            rating = rating_tag.text.strip() if rating_tag else '평가 없음'

            text_tag = comment.select_one(TEXT_SELECTOR)
            # 스포일러 '보기' 버튼이 클릭된 후에는 전체 텍스트가 포함되므로, '보기'라는 단어만 제거합니다.
            text = text_tag.text.strip().replace('보기', '') if text_tag else ''

            action_buttons = comment.select(ACTION_BUTTON_SELECTOR)
            likes = '0'
            replies = '0'
            for button in action_buttons:
                if '좋아요' in button.text:
                    likes = button.find('span').text if button.find('span') else '0'
                elif '댓글' in button.text:
                    replies = button.find('span').text if button.find('span') else '0'

            comments_data.append({
                "ID": user_id,
                "별점": rating,
                "내용": text,
                "좋아요": likes,
                "댓글 수": replies
            })

    except TimeoutException:
        print(f"[오류] 페이지 로딩 시간(20초)이 초과되었습니다. URL을 확인하거나 코드 상단의 COMMENT_ITEM_SELECTOR ('{COMMENT_ITEM_SELECTOR}')가 올바른지 확인해주세요.")
    except Exception as e:
        print(f"오류가 발생했습니다: {e}")
    finally:
        if driver:
            print("작업을 완료하고 브라우저를 종료합니다.")
            driver.quit()

    return comments_data

def print_guide():
    """콘텐츠 ID를 찾는 방법을 안내하는 함수"""
    print("=" * 60)
    print("왓챠피디아 콘텐츠 ID를 찾는 방법")
    print("=" * 60)
    print("1. 왓챠피디아에서 코멘트를 수집하고 싶은 영화/드라마 페이지로 접속합니다.")
    print("2. 브라우저의 주소창을 확인합니다.")

    print("   예시: 오징어 게임 시즌3(2025)로 코멘트가 비교적 많음. 대략 5분 이상 소요.)
    print("   예시: https://pedia.watcha.com/ko-KR/contents/tRMx0Jq/comments")
    print("3. 위 주소에서 'contents/' 와 '/comments' 사이에 있는")
    print("   영문과 숫자로 조합된 코드(예: tRMx0Jq)가 바로 '콘텐츠 ID' 입니다.")
    print("\n※ 안내: 코멘트가 많은 콘텐츠는 수집에 몇 분 이상 소요될 수 있습니다.")
    print("-" * 60)

if __name__ == "__main__":

    print_guide()

    content_id = ""
    while not content_id:
        content_id = input("▶ 수집할 콘텐츠의 ID를 입력하고 Enter 키를 누르세요: ").strip()
        if not content_id:
            print("[오류] ID가 입력되지 않았습니다. 다시 입력해주세요.")

    timeout_minutes = 0
    while True:
        timeout_input = input("▶ 최대 수집 시간을 분 단위로 입력하세요 (전체 수집은 0 또는 Enter): ").strip()
        if not timeout_input or timeout_input == "0":
            print("[알림] 시간 제한 없이 모든 댓글을 수집합니다.")
            break
        try:
            timeout_minutes = int(timeout_input)
            if timeout_minutes < 0:
                print("[오류] 0 또는 양수를 입력해주세요.")
            else:
                print(f"[알림] 최대 {timeout_minutes}분 동안 댓글을 수집합니다.")
                break
        except ValueError:
            print("[오류] 숫자만 입력해주세요.")

    comments = get_all_comments_with_selenium(content_id, timeout_minutes)

    if comments:
        print(f"\n총 {len(comments)}개의 코멘트를 성공적으로 수집했습니다.")
        file_name = f"watcha_comments_{content_id}.json"
        with open(file_name, "w", encoding="utf-8") as f:
            json.dump(comments, f, ensure_ascii=False, indent=4)
        print(f"'{file_name}' 파일로 저장이 완료되었습니다.")
    else:
        print("수집된 코멘트가 없습니다.")
````
